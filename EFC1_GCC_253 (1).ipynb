##  Insertion Sort


```python
def insertionSort(arranjo):

  comparacoes = [0]

  for j in range(1, len(arranjo)):
    chave = arranjo[j]
    i = j-1
    while i >= 0 and arranjo[i] > chave:
      comparacoes[0] += 1

      arranjo[i+1] = arranjo[i]
      i = i-1
      
    arranjo[i+1] = chave
    
  return arranjo, comparacoes[0]

A = [21, 12, 7, 27, 10, 1, 66, 2, 100, 43, 65, 34, 0, 56, 88, 8, 96, 91, 19, 30]

print(insertionSort(A))
```

##  Merge Sort


```python
def merge(arranjo, esquerda, meio, direita, aux, comparacoes):
    
    for x in range(esquerda, direita+1):
        aux[x] = arranjo[x]
    i = esquerda
    j = meio + 1
    
    for x in range(esquerda, direita+1):
        comparacoes[0] += 1
        if i > meio:
            arranjo[x] = aux[j]
            j += 1
        elif j > direita:
            arranjo[x] = aux[i]
            i += 1
        elif aux[j] < aux[i]:
            arranjo[x] = arranjo[j]
            j += 1
        else:
            arranjo[x] = aux[i]
            i += 1



def mergeSort(arranjo, esquerda, direita, aux):
    
    comparacoes = [0]

    if direita <= esquerda:
        return
    
    meio = (esquerda + direita) // 2
    
    mergeSort(arranjo, esquerda, meio, aux)
    
    mergeSort(arranjo, meio+1, direita, aux)
    
    merge(arranjo, esquerda, meio, direita, aux, comparacoes)

    return arranjo, comparacoes[0]

A = [21, 12, 7, 27, 10, 1, 66, 2, 100, 43, 65, 34, 0, 56, 88, 8, 96, 91, 19, 30]
aux = [0] * len(A)
print(mergeSort(A, 0, len(A) - 1, aux))
```

##  Selection Sort


```python
def selectionSort(arranjo):

  comparacoes = [0]
  tamanho = len(arranjo)
  for i in range(0, tamanho):
    min = i
    for j in range(i+1, tamanho):
      if arranjo[j] < arranjo[min]:
        min = j
        comparacoes[0] += 1
        
    if min != i:
      temp = arranjo[i]
      arranjo[i] = arranjo[min]
      arranjo[min] = temp
  
  return arranjo, comparacoes[0]

A = [21, 12, 7, 27, 10, 1, 66, 2, 100, 43, 65, 34, 0, 56, 88, 8, 96, 91, 19, 30]

print(selectionSort(A))
```

##  Bubble Sort


```python
def bubbleSort(arranjo):

  comparacoes = [0]
  n = len(arranjo) - 1
  alterado = True

  for i in range(0, n):
    if alterado == True :
      alterado = False

      for j in range(0, n-i):
        if arranjo[j] > arranjo[j+1]:
          arranjo[j+1], arranjo[j] = arranjo[j], arranjo[j+1]
          alterado = True

          comparacoes [0] += 1
    else:
        return arranjo, comparacoes[0]
  return arranjo, comparacoes[0]
  
  
A = [21, 12, 7, 27, 10, 1, 66, 2, 100, 43, 65, 34, 0, 56, 88, 8, 96, 91, 19, 30]

print(bubbleSort(A))
```

## Heap Sort

```python
def heapify(arranjo, tamanho, i, comparacoes):
    maior = i
    esquerda = 2 * i + 1
    direita = 2 * i + 2
    
    if esquerda < tamanho and arranjo[esquerda] > arranjo[maior]:
        comparacoes [0] += 1
        maior = esquerda
    
    if direita < tamanho and arranjo[direita] > arranjo[maior]:
        comparacoes [0] += 1
        maior = direita
        
    if maior != i:
        arranjo[maior], arranjo[i] = arranjo[i], arranjo[maior]
        heapify(arranjo, tamanho, maior, comparacoes)
        

def heapSort(arranjo):
    
    n = len(arranjo)
    comparacoes = [0]
    
    for i in range(n, -1, -1):
        heapify(arranjo, n, i, comparacoes)
        
    for i in range(n - 1, 0, -1):
        arranjo[i], arranjo[0] = arranjo[0], arranjo[i]
        heapify(arranjo, i, 0, comparacoes)

    return arranjo, comparacoes[0]


A = [21, 12, 7, 27, 10, 1, 66, 2, 100, 43, 65, 34, 0, 56, 88, 8, 96, 91, 19, 30]
print(heapSort(A))
```

## Quick Sort

> pivo = *A[A-comprimento]*

```python
def particinamento(arranjo, primeiro, ultimo, comparacoes):
    pivo = arranjo[ultimo]
    menor = primeiro

    for x in range(primeiro, ultimo):
      if arranjo[x] <= pivo:
        arranjo[menor], arranjo[x] = arranjo[x], arranjo[menor]
        menor = menor + 1
        comparacoes [0] += 1

    arranjo[menor], arranjo[ultimo] = arranjo[ultimo], arranjo[menor]
    
    return menor
    

def quickSort(arranjo, primeiro, ultimo):

    comparacoes = [0]
    if primeiro < ultimo:
      dividi = particinamento(arranjo, primeiro, ultimo, comparacoes)
      quickSort(arranjo, primeiro, dividi-1)
      quickSort(arranjo, dividi+1, ultimo)

    return arranjo, comparacoes[0]
    

A = [21, 12, 7, 27, 10, 1, 66, 2, 100, 43, 65, 34, 0, 56, 88, 8, 96, 91, 19, 30]

print(quickSort(A, 0, len(A) - 1))
```

###Gráfico-InsertionSort

O melhor caso acontece quando o vetor já esta ordenado.

O pior caso acontece quando o vetor esta ordenado de maneira contrária à ordenação que se deseja realizar;

ex.: se deseja ordenar de maneira crescente, então o pior caso seria o vetor ja estar ordenado de maneira decrescente, o mesmo acontece no caso contrário.

Melhor caso: $ O(n) $

Pior caso: $ {O(n^2)} $

Caso médio: $ {O(n^2)} $

```python
def tempoInsertion(A):
  inicio = time.time()
  insertionSort(A)
  fim = time.time()
  tempoFinal = fim -inicio
  return tempoFinal

yMedio = []
yMelhor = []
yPior =[]

ycomparaMedio = []
ycomparaMelhor = []
ycomparaPior = []

x = []

for t in range(5,1000,5):
 
  yMedio.append(tempoInsertion(np.random.randint(0, 1000, t)))
  yMelhor.append(tempoInsertion(np.arange(0,t,1)))
  yPior.append(tempoInsertion(np.arange(t,0,-1)))
  
  x.append(t)

  ycomparaMedio.append(tempoInsertion(np.random.randint(0, 1000, t)))
  ycomparaMelhor.append(tempoInsertion(np.arange(0,t,1)))
  ycomparaPior.append(tempoInsertion(np.arange(t,0,-1)))

  plt.suptitle("TEMPOS DE EXECUÇÃO INSERTION SORT")
  label2 = "NUMERO DE COMPARAÇÕES NO INSERTION SORT"

#Melhor caso: Verde
plt.plot(yMelhor, color='green')

#Caso médio: azul
plt.plot(yMedio, color='blue')

#Pior caso = vermelho
plt.plot(yPior, color='red')

#Legenda
plt.legend(["Melhor Caso","Caso Medio","Pior Caso"],title="Legenda")

#Grid
plt.grid(True)

#Labels
plt.xlabel("NÚMERO DE ENTRADAS")
plt.ylabel("T E M P O")

plt.show()


####
plt.suptitle(label2)

plt.plot(x,ycomparaMelhor, color='green')
plt.plot(x,ycomparaMedio,color='blue')
plt.plot(x,ycomparaPior, color='red')
  
plt.legend(["Crescente","aleatorio","Decrescente"],title="legenda")
plt.grid(True)
  
plt.xlabel("NÚMERO DE ENTRADAS")
plt.ylabel("COMPARAÇÕES")
plt.show()

```

###Gráfico-QuickSort

O melhor caso acontece quando o vetor não esta ordenado, possui os valores em ordem aleatória.

O pior caso ocorre quando o vetor já possui seus valores ordenados.

Melhor caso: $ O(nlogn) $

Pior caso: $ {O(n^2)} $

Caso médio: $ O(nlogn) $

```python
def tempoQuick(A):
  inicio = time.time()
  ultimo = len(A) - 1
  quickSort(A, 0, ultimo)
  fim = time.time()
  tempoFinal = fim -inicio
  return tempoFinal

yMelhor = []
yPior =[]

ycomparaMedio = []
ycomparaMelhor = []
ycomparaPior = []

x = []

for t in range(5,1000,5):
 
  yMelhor.append(tempoQuick(np.arange(0,t,1)))
  yPior.append(tempoQuick(np.arange(t,0,-1)))

  x.append(t)

  ycomparaMedio.append(tempoQuick(np.random.randint(0, 1000, t)))
  ycomparaMelhor.append(tempoQuick(np.arange(0,t,1)))
  ycomparaPior.append(tempoQuick(np.arange(t,0,-1)))
  
  plt.suptitle("TEMPOS DE EXECUÇÃO QUICK SORT COM O PIVO SENDO O ULTIMO ELEMENTO")
  label2 = "NUMERO DE COMPARAÇÕES NO QUICK SORT COM O PIVO SENDO O ULTIMO ELEMENTO"

#Melhor caso: Verde
plt.plot(yMelhor, color='green')

#Pior caso = vermelho
plt.plot(yPior, color='red')

#Legenda
plt.legend(["Melhor Caso","Pior Caso"],title="Legenda")

#Grid
plt.grid(True)

#Labels
plt.xlabel("NÚMERO DE ENTRADAS")
plt.ylabel("T E M P O")

plt.show()

####
plt.suptitle(label2)

plt.plot(x,ycomparaMelhor, color='green')
plt.plot(x,ycomparaMedio,color='blue')
plt.plot(x,ycomparaPior, color='red')
  
plt.legend(["Crescente","aleatorio","Decrescente"],title="legenda")
plt.grid(True)
  
plt.xlabel("NÚMERO DE ENTRADAS")
plt.ylabel("COMPARAÇÕES")
plt.show()
```

###Quick Sort Aleatório

```python
def partAle(arranjo, primeiro, ultimo, comparacoes):
    sorteio = np.random.randint(primeiro, ultimo)
    arranjo[ultimo], arranjo[sorteio] = arranjo[sorteio], arranjo[ultimo]
    
    pivo = arranjo[ultimo]
    menor = primeiro

    for x in range(primeiro, ultimo):
      if arranjo[x] <= pivo:
        arranjo[menor], arranjo[x] = arranjo[x], arranjo[menor]
        menor = menor + 1
        comparacoes [0] += 1

    arranjo[menor], arranjo[ultimo] = arranjo[ultimo], arranjo[menor]
    
    return menor
    

def quickSortAle(arranjo, primeiro, ultimo):

    comparacoes = [0]
    if primeiro < ultimo:
      dividi = partAle(arranjo, primeiro, ultimo, comparacoes)
      quickSortAle(arranjo, primeiro, dividi-1)
      quickSortAle(arranjo, dividi+1, ultimo)

    return arranjo, comparacoes[0]
    

A = [21, 12, 7, 27, 10, 1, 66, 2, 100, 43, 65, 34, 0, 56, 88, 8, 96, 91, 19, 30]

print(quickSortAle(A, 0, len(A) - 1))
```
